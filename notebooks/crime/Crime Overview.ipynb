```python
%pylab inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import ambry
import pandas as pd

ab = ambry.new_analysis_bundle(source='sandiegodata.org', dataset='crime',
                        subset='test', bspace='casnd', revision=1)
```

Set some configuration for the bundle. 

```python
with ab.session as s:
    ab.partitions.clean(s)

with ab.config.about as a:
    a.title = 'An Analysis Bundle Example Using Crime Data'
    a.tags = ['tag1','tag2']
    a.groups = ['group1']

ab.config.build.dependencies =  {
    'incidents' : 'clarinova.com-crime-incidents-casnd-linked-crimes',
    'addresses' : 'clarinova.com-crime-incidents-casnd-addresses',
    'places' : 'clarinova.com-places-casnd-areastats-0.1.6'
}
```

```python
incidents = ab.library.dep('incidents').partition
addresses = ab.library.dep('addresses').partition
addresses = ab.library.dep('places').partition

incidents.attach(addresses, 'addr')

```

```python
from datetime import datetime
df =  incidents.select("SELECT * FROM crimes LIMIT 500000",index_col='id').pandas
df.count()
```

```python
dfm= df.groupby(['hour','legend'])
counts = dfm.count()
cus = counts.unstack(1)
uscounts = cus['id']
ax = uscounts.plot(subplots=True, figsize=(10,20), colormap='jet')
```

```python
dowg = df.groupby(['dow','legend'])
dowcounts = dowg.count()
dowuns = dowcounts.unstack(1)
dowcounts = dowuns['id']
ax = dowcounts.plot(subplots=True, figsize=(10,20))
```

```python
from math import ceil

legends = [ l for l in df.legend.unique() if l.strip() and l not in ('-', 'ARSON','HOMICIDE') ]

c = len(legends)
ncols = 3.
nrows = ceil(float(c)/ncols)
print nrows, ncols

cmap = plt.get_cmap('YlOrRd')

fig, axes = plt.subplots(nrows=int(nrows), ncols=int(ncols), figsize=figsize(10,10), squeeze=True, sharex=False, sharey=False)

fig. set_size_inches(10,10)
axes = axes.ravel() # Convert 2D array to 1D
plt.tight_layout(h_pad=1)
for i, legend in enumerate(legends):

    sub = df[df.legend == legend]
    heatg = sub.groupby(['dow','hour'])
    hgcounts = heatg.count()['id'].unstack('hour')
   

    # Converting to an array puts it into a for that
    # matplotlib expects. This probably only works b/c the
    # hours and days of week are 1-based indexes. 
    axes[i].pcolormesh(array(hgcounts.T),cmap=cmap)
    axes[i].set_title(legend.title())
    
```
